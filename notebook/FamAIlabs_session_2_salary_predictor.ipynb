# **ML - Model implement**

Before jumping into Linear Regression, students need a simple mental framework.

The easiest way to explain the core of Machine Learning is through Learning with Examples versus Learning by Finding Patterns

***Core Concept how Machine learns:***

1.   **Supervised Learning**: Learning with an Answer Key (The model gets both the questions and the answers)
2.   **Unsupervised Learning**: Learning by Grouping (The model gets only the data and must find its own patterns)

**Data Prep -**

````python
X = df_salary[['YearsExperience']]  # my Inputr feature - here it's only single column / in future we will add more columns - 2 dimenisional array
y = df_salary['Salary']  # target - always single column  - 1 dimensional array

print(y)

print(X)
````

**Model -Tarin-test**

###*The 3-Step Model Training Workflow*

```text
       ┌───────────────┐        ┌──────────────┐        ┌──────────────┐
       │  INPUT DATA   │        │   AI MODEL   │        │ PREDICTIONS  │
       │   (X & y)     │ ─────► │  (Training)  │ ─────► │  (The Rule)  │
       │               │        │  .fit(X, y)  │        │  .predict()  │
       └───────────────┘        └──────────────┘        └──────────────┘
    [Years vs. Salary]       [Studies Patterns]       [Calculates Future]
```
* **Step 1:** Feed the model your historical experience and salary records.
* **Step 2:** Run `.fit(X, y)` to let the algorithm study the patterns and learn the trend line.
* **Step 3:** Use `.predict()` to input brand-new experience numbers and get instant salary estimations.


**Model Note** : Why Use Linear Regression Here?We use Linear Regression because our goal is to predict a continuous numeric value (Salary) based on a straight-line relationship with our input (YearsExperience). As experience goes up, salary steadily goes up in a predictable line.

````python

'''Scikit learn (sklearn) is the datascience tool kit in Pytho and we're importing alreday written
 LinearRegression ML model as mentioned in the model note (as x and y are proportionate)'''

from sklearn.linear_model import LinearRegression

''' instantitate the model - We create a brand-new,
completely blank virtual robot brain that is ready and waiting to start learning from our data.'''
model = LinearRegression()

''' Model fit on X and y data - feature and target'''
model.fit(X, y)

# The model learns relationship patterns from historical data
````

**Make Predictions**

````python
prediction = model.predict([[9]])

print(prediction)  # prediction of salary of 9 years expericne
````

# **SHOW BEST FIT LINE**

````python
plt.scatter(df_salary['YearsExperience'], df_salary['Salary'])

plt.plot(df_salary['YearsExperience'], model.predict(X),color='g')  # green color best fit line as we use color attribute

plt.xlabel("Years of Experience")
plt.ylabel("Salary")
plt.title("Salary Prediction Line")

print(" ****  The model tries to draw the best-fit relationship using historical data :) ***")
plt.show()
````
